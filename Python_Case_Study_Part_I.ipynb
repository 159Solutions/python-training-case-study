```python
Master_df = final_pivot_2.merge(Selected_Part_D,on='NPI',how='outer').merge(Selected_Open_Payments_df,on='NPI',how='outer').merge(\
            NPPES_df_deduped,on='NPI',how='left').merge(Sub_Spec_df,on='NPI',how='left')
print(Master_df.count())
Master_df[Master_df['NPI']=="1003017971"].head(100)
```

```python
for col in Master_df.columns:
    if ("VOLUME" in col) or ("COUNT" in col) or ("PAYMENT" in col):
        Master_df[col].fillna(0,inplace = True)
        
```

```python
## Create decile function which creates a new decile column for each measure e.g. CLAIM_VOLUME_P
## decile column should have zero value for zeroes/NA in measure columns
## Also create decile ranks from 1-10 for non-zero measure value based on the respective mesure column
## e.g. decile column for CLAIM_VOLUME_P
Decile_Master_df = Master_df
for col in Master_df.columns:
    if "VOLUME" in col:
        Decile_Master_df = Decile_Master_df.sort_values(by=col,ascending=False)
        Decile_Master_df['cumm_vol'] = Decile_Master_df[col].cumsum()
        max=Decile_Master_df.iloc[-1,Decile_Master_df.columns.get_loc('cumm_vol')]/10
        
        Decile_Master_df['DECILE'+'_'+col]=10-np.floor(Decile_Master_df['cumm_vol']/max)
        Decile_Master_df=Decile_Master_df.drop('cumm_vol',axis=1)
print(Decile_Master_df.head(10))
```

```python
Selected_Decile_Master_df = Decile_Master_df.filter(['NPI'] + [col for col in Decile_Master_df.columns if "CLAIM" in col or "PATIENT" in col], axis=1)
Selected_Decile_Master_df.corr()
```

```python
Relative_df = Selected_Decile_Master_df.groupby('DECILE_CLAIM_VOLUME_P').mean().reset_index().round(1)
Relative_df
```

```python
## Calculate scaled score for each decile for each metric bout of 1000
Total_Relative_df = pd.DataFrame()
Total_Relative_df['DECILE'] = Selected_Decile_Master_df.sort_values(by='DECILE_CLAIM_VOLUME_P',ascending=False)['DECILE_CLAIM_VOLUME_P'].drop_duplicates()
for col in Selected_Decile_Master_df.columns:
    if 'DECILE' in col:
        new_col = col.replace('DECILE_', '')
        Relative_df = Selected_Decile_Master_df.groupby(col)[new_col].mean().reset_index().round(1)
        Relative_df['SCALED_'+new_col] = Relative_df[new_col]*1000/Relative_df.iloc[-1,Relative_df.columns.get_loc(new_col)]
        Total_Relative_df = Total_Relative_df.merge(Relative_df, \
                                left_on = 'DECILE', right_on = col, how='left' ).drop(col, axis = 1).drop(new_col, axis = 1)
        
Total_Relative_df = Total_Relative_df.round(2)
Total_Relative_df
```

```python
## Get scaled scores for each NPI based on each measure
for col in Selected_Decile_Master_df.columns:
    if ('DECILE' in col):
        new_col = col.replace('DECILE_','')
        Selected_Decile_Master_df= Selected_Decile_Master_df.merge(Total_Relative_df[['DECILE','SCALED_'+ new_col]],\
                                    left_on = col, right_on = 'DECILE',how = 'left' ).drop('DECILE',axis=1)
Selected_Decile_Master_df.head()
```
